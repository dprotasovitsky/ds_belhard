 –ù–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö –æ –∂–∏–ª—å–µ –≤ –ö–∞–ª–∏—Ñ–æ—Ä–Ω–∏–∏ (California housing dataset), –∫–æ—Ç–æ—Ä—ã–π –≤–∫–ª—é—á–µ–Ω –≤ scikit-learn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import joblib
from data_loader import AdvancedDataLoader
from model_trainer import AdvancedModelTrainer
from ensemble_model import AdvancedEnsembleModel
from visualization import AdvancedVisualization
from config import Config


def setup_environment():
    """–ù–∞—Å—Ç—Ä–æ–π–∫–∞ –æ–∫—Ä—É–∂–µ–Ω–∏—è"""
    print("\U0001f680 –ù–ê–°–¢–†–û–ô–ö–ê –û–ö–†–£–ñ–ï–ù–ò–Ø –î–õ–Ø –ê–ù–ê–õ–ò–ó–ê –†–ï–ì–†–ï–°–°–ò–ò")
    print("=" * 70)

    # –°–æ–∑–¥–∞–Ω–∏–µ –ø–∞–ø–æ–∫ –¥–ª—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    os.makedirs(Config.MODEL_SAVE_PATH, exist_ok=True)
    os.makedirs(Config.RESULTS_SAVE_PATH, exist_ok=True)

    # –£—Å—Ç–∞–Ω–æ–≤–∫–∞ —Å–ª—É—á–∞–π–Ω–æ–≥–æ seed
    Config.set_random_seed()

    print("\U00002705 –û–∫—Ä—É–∂–µ–Ω–∏–µ –Ω–∞—Å—Ç—Ä–æ–µ–Ω–æ")


def main():
    """–û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è"""
    setup_environment()

    # –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–æ–≤
    data_loader = AdvancedDataLoader()
    model_trainer = AdvancedModelTrainer()
    ensemble_creator = AdvancedEnsembleModel()
    visualizer = AdvancedVisualization(style="seaborn")

    # –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö
    print("\n\U0001f4ca –ó–ê–ì–†–£–ó–ö–ê –ò –ê–ù–ê–õ–ò–ó –î–ê–ù–ù–´–•")
    print("=" * 70)

    # –í—ã–±–æ—Ä –¥–∞—Ç–∞—Å–µ—Ç–∞
    dataset_type = "california"  # –ú–æ–∂–Ω–æ –∏–∑–º–µ–Ω–∏—Ç—å –Ω–∞ 'synthetic' –∏–ª–∏ 'synthetic_complex'
    df, feature_names = data_loader.load_dataset(dataset_type)
    target_column = "PRICE" if "PRICE" in df.columns else "target"

    # –ê–Ω–∞–ª–∏–∑ –¥–∞—Ç–∞—Å–µ—Ç–∞
    data_loader.analyze_dataset(df, target_column)

    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –¥–∞—Ç–∞—Å–µ—Ç–∞
    visualizer.plot_dataset_analysis(df, target_column)

    # –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö
    print("\n\U0001f527 –ü–û–î–ì–û–¢–û–í–ö–ê –î–ê–ù–ù–´–•")
    print("=" * 70)
    X_train, X_test, y_train, y_test, features = data_loader.prepare_data(
        df, target_column
    )

    print(f"–†–∞–∑–º–µ—Ä train: {X_train.shape}")
    print(f"–†–∞–∑–º–µ—Ä test: {X_test.shape}")
    print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤: {len(features)}")
    # –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π
    results = model_trainer.train_all_models(
        X_train,
        X_test,
        y_train,
        y_test,
        feature_names=features,
        tune=True,
        tuning_method="random",  # –ú–æ–∂–Ω–æ –∏–∑–º–µ–Ω–∏—Ç—å –Ω–∞ 'grid'
    )

    # –í—ã–±–æ—Ä –ª—É—á—à–∏—Ö –º–æ–¥–µ–ª–µ–π –¥–ª—è –∞–Ω—Å–∞–º–±–ª—è
    print("\n\U0001f3af –í–´–ë–û–† –õ–£–ß–®–ò–• –ú–û–î–ï–õ–ï–ô –î–õ–Ø –ê–ù–°–ê–ú–ë–õ–Ø")
    print("=" * 70)
    best_models = model_trainer.select_best_models(results, n_models=3, metric="rmse")

    # –°–æ–∑–¥–∞–Ω–∏–µ –∏ –æ—Ü–µ–Ω–∫–∞ –∞–Ω—Å–∞–º–±–ª–µ–π
    print("\n\U0001f91d –°–û–ó–î–ê–ù–ò–ï –ê–ù–°–ê–ú–ë–õ–ï–ô")
    print("=" * 70)
    ensembles = ensemble_creator.create_ensembles(best_models)
    ensemble_results = ensemble_creator.evaluate_ensembles(
        ensembles, X_train, X_test, y_train, y_test
    )

    # –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –∞–Ω—Å–∞–º–±–ª–µ–π —Å –æ–¥–∏–Ω–æ—á–Ω—ã–º–∏ –º–æ–¥–µ–ª—è–º–∏
    ensemble_creator.compare_with_single_models(ensemble_results, results)

    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    print("\n\U0001f4c8 –í–ò–ó–£–ê–õ–ò–ó–ê–¶–ò–Ø –†–ï–ó–£–õ–¨–¢–ê–¢–û–í")
    print("=" * 70)

    # –°–æ–∑–¥–∞–Ω–∏–µ DataFrame —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏
    results_df = model_trainer.get_results_dataframe()

    # –î–æ–±–∞–≤–ª—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –∞–Ω—Å–∞–º–±–ª–µ–π
    ensemble_data = []
    for name, result in ensemble_results.items():
        ensemble_data.append(
            {
                "Model": f"ENSEMBLE_{name}",
                "Train_RMSE": f"{result['train_rmse']:.4f}",
                "Test_RMSE": f"{result['test_rmse']:.4f}",
                "Train_R2": f"{result['train_r2']:.4f}",
                "Test_R2": f"{result['test_r2']:.4f}",
                "Test_MAE": "-",
                "Test_MAPE": "-",
                "CV_RMSE": "-",
                "CV_Std": "-",
                "Training_Time": "-",
            }
        )

    all_results_df = pd.concat(
        [results_df, pd.DataFrame(ensemble_data)], ignore_index=True
    )

    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è –º–æ–¥–µ–ª–µ–π
    visualizer.plot_model_comparison(all_results_df)

    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –ª—É—á—à–∏—Ö –º–æ–¥–µ–ª–µ–π
    best_single_name = min(results.items(), key=lambda x: x[1]["test_metrics"]["rmse"])[
        0
    ]
    best_ensemble_name = min(ensemble_results.items(), key=lambda x: x[1]["test_rmse"])[
        0
    ]

    y_pred_dict = {
        best_single_name: results[best_single_name]["y_test_pred"],
        best_ensemble_name: ensemble_results[best_ensemble_name]["y_test_pred"],
    }

    visualizer.plot_predictions_comparison(
        y_test, y_pred_dict, [best_single_name, best_ensemble_name]
    )

    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≤–∞–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è –º–æ–¥–µ–ª–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –µ–µ –ø–æ–¥–¥–µ—Ä–∂–∏–≤–∞—é—Ç
    feature_importance_dict = {}
    for model_name, importance in model_trainer.feature_importance.items():
        if np.sum(importance) > 0:  # –ü—Ä–æ–≤–µ—Ä—è–µ–º, —á—Ç–æ –≤–∞–∂–Ω–æ—Å—Ç—å –Ω–µ –Ω—É–ª–µ–≤–∞—è
            feature_importance_dict[model_name] = importance

    if feature_importance_dict:
        visualizer.plot_feature_importance(feature_importance_dict, features, top_n=10)

    # –ò–Ω—Ç–µ—Ä–∞–∫—Ç–∏–≤–Ω—ã–µ –≥—Ä–∞—Ñ–∏–∫–∏
    print("\n\U0001f4ca –ò–ù–¢–ï–†–ê–ö–¢–ò–í–ù–´–ï –ì–†–ê–§–ò–ö–ò")
    visualizer.create_interactive_plot(
        y_test, results[best_single_name]["y_test_pred"], best_single_name
    )

    # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    print("\n\U0001f4be –°–û–•–†–ê–ù–ï–ù–ò–ï –†–ï–ó–£–õ–¨–¢–ê–¢–û–í")
    print("=" * 70)

    # –û–ø—Ä–µ–¥–µ–ª—è–µ–º, —á—Ç–æ —Å–æ—Ö—Ä–∞–Ω—è—Ç—å: –ª—É—á—à—É—é –æ–¥–∏–Ω–æ—á–Ω—É—é –º–æ–¥–µ–ª—å –∏–ª–∏ –∞–Ω—Å–∞–º–±–ª—å
    best_single_rmse = results[best_single_name]["test_metrics"]["rmse"]
    best_ensemble_rmse = ensemble_results[best_ensemble_name]["test_rmse"]

    if best_ensemble_rmse < best_single_rmse:
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º –∞–Ω—Å–∞–º–±–ª—å
        ensemble_creator.save_best_ensemble(ensemble_results, "best_ensemble_model.pkl")
        final_model_type = "–∞–Ω—Å–∞–º–±–ª—å"
        final_model_name = best_ensemble_name
    else:
        # –°–æ—Ö—Ä–∞–Ω—è–µ–º –æ–¥–∏–Ω–æ—á–Ω—É—é –º–æ–¥–µ–ª—å
        model_trainer.save_best_model(results, "best_single_model.pkl")
        final_model_type = "–æ–¥–∏–Ω–æ—á–Ω–∞—è –º–æ–¥–µ–ª—å"
        final_model_name = best_single_name

    # –°–æ—Ö—Ä–∞–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –≤ CSV
    all_results_df.to_csv(
        f"{Config.RESULTS_SAVE_PATH}model_comparison.csv", index=False
    )
    print(
        f"\U0001f4cb –†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤ {Config.RESULTS_SAVE_PATH}model_comparison.csv"
    )

    # –§–∏–Ω–∞–ª—å–Ω—ã–π –æ—Ç—á–µ—Ç
    print("\n\U0001f389 –§–ò–ù–ê–õ–¨–ù–´–ô –û–¢–ß–ï–¢")
    print("=" * 70)
    print(f"–õ—É—á—à–∞—è {final_model_type}: {final_model_name}")
    print(f"–õ—É—á—à–∏–π RMSE: {min(best_single_rmse, best_ensemble_rmse):.4f}")
    print(f"–û–±—É—á–µ–Ω–æ –º–æ–¥–µ–ª–µ–π: {len(results)}")
    print(f"–ü—Ä–æ—Ç–µ—Å—Ç–∏—Ä–æ–≤–∞–Ω–æ –∞–Ω—Å–∞–º–±–ª–µ–π: {len(ensemble_results)}")
    print(f"–ú–æ–¥–µ–ª–∏ —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤: {Config.MODEL_SAVE_PATH}")
    print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å–æ—Ö—Ä–∞–Ω–µ–Ω—ã –≤: {Config.RESULTS_SAVE_PATH}")

    # –ê–Ω–∞–ª–∏–∑ —Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏—Ö –¥–∞–Ω–Ω—ã—Ö
    if dataset_type.startswith("synthetic"):
        print("\n\U0001f52c –î–û–ü–û–õ–ù–ò–¢–ï–õ–¨–ù–´–ô –ê–ù–ê–õ–ò–ó –°–ò–ù–¢–ï–¢–ò–ß–ï–°–ö–ò–• –î–ê–ù–ù–´–•")
        print("=" * 70)

    # –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω—ã—Ö —Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏—Ö –¥–∞–Ω–Ω—ã—Ö –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è
    synthetic_loader = AdvancedDataLoader()
    synthetic_df, synthetic_features = synthetic_loader.load_dataset(
        "synthetic_complex"
    )

    print("–°–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ —Å–ª–æ–∂–Ω—ã–µ —Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ:")
    print(f"–†–∞–∑–º–µ—Ä: {synthetic_df.shape}")
    print(f"–û—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–∏: {len(synthetic_features)} –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Å —Ä–∞–∑–Ω—ã–º–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è–º–∏")
    print(f"–¶–µ–ª–µ–≤–∞—è –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è: {synthetic_df['target'].describe()}")

    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Å–ª–æ–∂–Ω—ã—Ö —Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏—Ö –¥–∞–Ω–Ω—ã—Ö
    visualizer.plot_dataset_analysis(synthetic_df, "target")


if __name__ == "__main__":
    main()
```
